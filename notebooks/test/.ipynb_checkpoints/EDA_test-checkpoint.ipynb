```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\User\Documents\horse-racing-arima-kinen-2013\data\feature_engineer_data.xlsx")
```

```python
horse_distance = (
    df.groupby(['Horse', 'Distance'])['Finish']
      .mean()
      .reset_index()
      .sort_values(['Horse', 'Distance'])
)
```

```python
# Get unique horse names
horses = horse_distance['Horse'].unique()
n_horses = len(horses)

# Plot 4 horses per figure
for i in range(0, n_horses, 4):
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    axes = axes.flatten()

    for j, ax in enumerate(axes):
        if i + j < n_horses:
            horse = horses[i + j]
            data = horse_distance[horse_distance['Horse'] == horse]
            ax.plot(data['Distance'], data['Finish'], marker='o', label=horse)
            ax.set_title(horse)
            ax.set_xlabel("Distance (m)")
            ax.set_ylabel("Avg Finish")
            ax.invert_yaxis()  # Lower finish = better, so invert
            ax.legend()

    plt.tight_layout()
    plt.suptitle(f"Average Finish by Distance (Horses {i+1}–{min(i+4, n_horses)})", y=1.02)
    plt.show()
```

```python
# Compute global average per distance
global_avg = df.groupby('Distance')['Finish'].mean()

# Merge into horse-distance DataFrame
horse_distance['GlobalAvg'] = horse_distance['Distance'].map(global_avg)

# Difference from global average (negative = better)
horse_distance['DiffFromAvg'] = horse_distance['Finish'] - horse_distance['GlobalAvg']
```

```python
# Plot 4 horses per figure
for i in range(0, n_horses, 4):
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    axes = axes.flatten()

    for j, ax in enumerate(axes):
        if i + j < n_horses:
            horse = horses[i + j]
            data = horse_distance[horse_distance['Horse'] == horse]
            ax.plot(data['Distance'], data['DiffFromAvg'], marker='o', label=horse)
            ax.axhline(0, color='gray', linestyle='--', linewidth=1)
            ax.set_title(horse)
            ax.set_xlabel("Distance (m)")
            ax.set_ylabel("Finish vs Global Avg")
            ax.legend()
    
    plt.tight_layout()
    plt.suptitle(f"Performance vs Global Avg by Distance (Horses {i+1}–{min(i+4, n_horses)})", y=1.02)
    plt.show()
```

```python
pivot = horse_distance.pivot(index='Horse', columns='Distance', values='Finish')
plt.figure(figsize=(10,6))
plt.imshow(pivot, cmap='RdYlGn_r', aspect='auto')
plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=45)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.colorbar(label='Average Finish Position')
plt.title('Horse Performance by Distance')
plt.show()
```

```python
df['Speed_mps'] = df['Distance'] / df['FT_seconds']  # meters per second
df['Speed_kph'] = df['Speed_mps'] * 3.6  # convert to km/h
```

```python
horse_speed = (
    df.groupby(['Horse', 'Distance'])['Speed_kph']
      .mean()
      .reset_index()
)
```

```python
# Get list of all unique horses
horses = horse_speed['Horse'].unique()

# Loop through each horse and plot individually
for horse in horses:
    data = horse_speed[horse_speed['Horse'] == horse]

    plt.figure(figsize=(7, 5))
    plt.plot(data['Distance'], data['Speed_kph'], marker='o', linestyle='-', label=horse)

    plt.title(f"{horse} – Speed vs Distance")
    plt.xlabel("Distance (m)")
    plt.ylabel("Average Speed (km/h)")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
```
